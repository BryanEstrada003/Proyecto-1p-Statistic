```python
import random as rd
def calcularGanancias(retweets):
    ganancias = retweets * rd.randint(3,5)
    return ganancias

def popularidad(fila):
    result = fila["followees"]/fila["followers"]
    return result

df["ganancias"] = df["retweets"].apply(calcularGanancias)
df["popularidad"] = df.apply(popularidad, axis = 1) #axis = 1 indica que la funcion debe ser aplicada por cada fila del df
df.head()
```

```python
df.groupby("country").mean()
```

```python
group = df.groupby("country").agg({
    "followers": "sum",
    "mentions": "mean",
    "retweets": "max",
})

group[group["followers"]>5000]
```

```python
group.to_csv("salida.csv")
```

# Pandas II - Join

```python
df1 = pd.read_csv("dataset_2a.csv")
df2 = pd.read_csv("dataset_2b.csv")
df3 = pd.read_csv("dataset_2c.csv")
```

```python
df1.head()
```

```python
df2
```

```python
df3
```

```python
df_with_users=df1.merge(df2, on = "user", how="left") # join
df_with_users.head()
```

```python
df_complete=df_with_users.merge(df3, on= "ciudad", how="left")
df_complete.head()
```

```python
df_complete["fecha"] = pd.to_datetime(df_complete["fecha_publicado"], format="%d/%m/%Y")
df_complete.head()
```

```python
df_complete["dia_semana"] = df_complete["fecha"].apply(lambda x: x.strftime("%A")) #dia de la semana
df_complete.head()
```

```python
df_complete["mes"] = df_complete["fecha_publicado"].apply(lambda x: x[3:]) #obtener mes y año
df_complete.head()
```

# PANDAS III - Agrupaciones múltiples, pivotar y transponer

```python
agrupado = df_complete.groupby(["mes", "ciudad"]).agg({
    "id": "count"
})
agrupado # guayaquil y mes ahora son indices
```

```python
desagrupar = agrupado.reset_index(1) #el indice de ciudad ahora es una columna
desagrupar
```

```python
desagrupar[["ciudad"]]
```

```python
#pivotar: cada valor unico se vuelve una nueva columna
pivote = desagrupar.pivot(columns="ciudad")
pivote
```

```python
pivote = pivote.droplevel(0, axis="columns")## tiene dos niveles, eliminamos el id
pivote
```

```python
pivote[["GUAYAQUIL"]]
```

```python
pivote = pivote.fillna(0)
pivote
```

```python
# transponer: de fila a columna y viceversa
pivote.transpose()
```
